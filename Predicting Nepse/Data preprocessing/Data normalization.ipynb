```python
import pandas as pd
```

### Removing unnecessary dates

```python
# Load the merged data
merged_data = pd.read_csv('merged data.csv')  # Update with the correct file path
merged_data
```

```python
min_val = merged_data['Close'].min()
max_val = merged_data['Close'].max()
print(min_val)
print(max_val)
```

```python
# Remove the specified columns
columns_to_remove = ['Date (AD)', 'Date (BS)', 'Date', 'Year', 'Start_Date_AD', 'End_Date_AD']
merged_data_cleaned = merged_data.drop(columns=columns_to_remove, errors='ignore')
```

```python
merged_data_cleaned.columns
```

```python
# Load the merged data
merged_data = pd.read_csv('merged data.csv')  

min_val = merged_data['Close'].min()
max_val = merged_data['Close'].max()

# Normalizing the numeric columns using Min-Max Scaling
for column in merged_data_cleaned.select_dtypes(include=['float64', 'int64']).columns:
    if column != 'Month':
        min_val = merged_data_cleaned[column].min()
        max_val = merged_data_cleaned[column].max()
        merged_data_cleaned[column] = (merged_data_cleaned[column] - min_val) / 
        (max_val - min_val) if max_val != min_val else merged_data_cleaned[column]
        
# Create 12 new columns for each month (1 for Baisakh, 2 for Jestha, etc.)
month_names = ['Baisakh', 'Jestha', 'Ashar', 'Shrawan', 'Bhadra', 'Asoj', 'Kartik', 
               'Mangsir', 'Poush', 'Magh', 'Falgun', 'Chaitra']
for month in month_names:
    merged_data_cleaned[month] = (merged_data_cleaned['Month'] == (month_names.index(month) + 1)).astype(int)
merged_data_cleaned
```

```python
# Create 12 new columns for each month (1 for Baisakh, 2 for Jestha, etc.)
month_names = ['Baisakh', 'Jestha', 'Ashar', 'Shrawan', 'Bhadra', 'Asoj', 'Kartik', 'Mangsir', 'Poush', 'Magh', 'Falgun', 'Chaitra']
for month in month_names:
    merged_data_cleaned[month] = (merged_data_cleaned['Month'] == (month_names.index(month) + 1)).astype(int)
merged_data_cleaned
```
