```python
import pandas as pd
```

```python
# Load your data
df = pd.read_csv('final_data.csv')
df
```

```python
# Create 60-day windows with the 61st day's Close as output
for i in range(len(df) - 60):
    input_data = df.iloc[i:i+60].values.flatten()  # Include 'Close' in input
    output_data = df.iloc[i+60]['Close']           # 61st day's closing price
    organized_row = list(input_data) + [output_data]
    organized_data.append(organized_row)

# Generate column names for each feature across 60 days
column_names = [f"{feature}_Day{day}" for day in range(1, 61) for feature in df.columns] + ['Output']

# Convert to a DataFrame
organized_df = pd.DataFrame(organized_data, columns=column_names)

# Save the organized data
organized_df.to_csv('organized_final_data.csv', index=False)
```
