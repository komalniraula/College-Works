```python
import pandas as pd
import nepali_datetime
import datetime
```

### Convert AD to BS

```python
def ad_to_bs(date_string):
    # Parse the date string into a datetime object
    ad_date = datetime.datetime.strptime(date_string, '%m/%d/%Y')

    # Convert to BS date
    bs_date = nepali_datetime.date.from_datetime_date(ad_date.date())

    # Format the date in a readable string format
    return bs_date.strftime('%Y-%m-%d')

# Load the CSV file
file_path = 'NEPSEindex.csv'  # Replace with your file path
nepse_data = pd.read_csv(file_path)

# Convert all AD dates to BS and store in a new column
nepse_data['Date (BS)'] = nepse_data['Date'].apply(ad_to_bs)

nepse_data
```

### Convert BS month to AD range

```python
import pandas as pd
import nepali_datetime

# Load the datasets
calendar_bs_path = 'calendar_bs.csv'  
monthly_data_path = 'monthly data.xlsx'  

# Read the datasets
calendar_bs = pd.read_csv(calendar_bs_path)
monthly_data_bs = pd.read_excel(monthly_data_path)

# Split the 'Date' column into 'Year' and 'Month' in the monthly data
monthly_data_bs[['Year', 'Month']] = monthly_data_bs['Date'].str.split('/', expand=True).astype(int)

# Define a function to get the Gregorian start and end dates for a BS month
def get_ad_dates(year, month, calendar):
    # Mapping BS month number to calendar_bs column names
    month_names = {
        1: "Baisakh", 2: "Jestha", 3: "Ashar", 4: "Shrawan", 5: "Bhadra",
        6: "Asoj", 7: "Kartik", 8: "Mangsir", 9: "Poush", 10: "Magh",
        11: "Falgun", 12: "Chait"
    }

    # Get the number of days in the BS month from the calendar
    days_in_month = calendar[calendar['Year'] == year][month_names[month]].iloc[0]

    # Convert the start and end of the BS month to Gregorian dates
    start_bs = nepali_datetime.date(year, month, 1)
    start_ad = start_bs.to_datetime_date()
    end_bs = nepali_datetime.date(year, month, days_in_month)
    end_ad = end_bs.to_datetime_date()

    return start_ad, end_ad

# Apply the function to each row in the monthly data
monthly_data_bs['Start_Date_AD'], monthly_data_bs['End_Date_AD'] = zip(*monthly_data_bs.apply(lambda row: get_ad_dates(row['Year'], row['Month'], calendar_bs), axis=1))
```

```python
# Save the updated dataframe to a new CSV file
monthly_data_bs.to_csv('monthly data.csv', index=False)
```
